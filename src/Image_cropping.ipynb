```python
from PIL import Image
from itertools import product
import os
import matplotlib.pyplot as plt
import numpy as np
import time
import pandas as pd
from matplotlib.image import imread

master_path = '/Users/angelazhang/Desktop/A/'
classes = pd.read_csv(os.path.join(master_path, "class_dict.csv"))
annotations = pd.read_csv(os.path.join(master_path, "metadata.csv"))

def tile(filename, dir_in, dir_out, d, ide, ty):
    name, ext = os.path.splitext(filename)
    img = Image.open(os.path.join(dir_in, filename))
    w, h = img.size
    path = []
    ID = []
    TY = []
    grid = product(range(0, h-h%d, d), range(0, w-w%d, d))
    x = 0
    for i, j in grid:
        box = (j, i, j+d, i+d)
        out = os.path.join(dir_out, f'{name}_{x}{ext}')
        path.append(f'{name}_{x}{ext}')
        ID.append(ide)
        TY.append(ty)
        #img.crop(box).save(out)
        x += 1
    return path, ID, TY

dir_in = '/Users/angelazhang/Desktop/A/'
dir_out = '/Users/angelazhang/Desktop/croped_dataset/'
d = 612

sat_crop_path = []
TY = []
ID = []
for i in range(len(annotations['mask_path'])):
  filename = annotations['sat_image_path'][i]
  ide = annotations['image_id'][i]
  ty = annotations['split'][i]
  sat_crop_path_new, ID_new, TY_new = tile(filename, dir_in, dir_out, d, ide, ty)
  sat_crop_path = sat_crop_path + sat_crop_path_new
  ID = ID + ID_new
  TY = TY + TY_new
  print(i)

mask_crop_path = []
# len(annotations['mask_path'])
for i in range(len(annotations['mask_path'])):
  filename = annotations['mask_path'][i]
  ide = annotations['image_id'][i]
  ty = annotations['split'][i]
  mask_crop_path_new = tile(filename, dir_in, dir_out, d, ide, ty)[0]
  mask_crop_path = mask_crop_path + mask_crop_path_new
  print(i)


df = pd.DataFrame()
df['image_id'] = ID
df['split'] = TY
df['sat_image_path'] = sat_crop_path
df['mask_path'] = mask_crop_path



df.to_csv('/Users/angelazhang/Desktop/A/croped_image.csv', index=False)
```
